```python
import pandas as pd
```

```python
# Files to load
file_to_load = './Resources/missing_grades.csv'

# Read the CSV into a DataFrame
missing_grade_df = pd.read_csv(file_to_load)
missing_grade_df
```

```python
# Drop the NaNs.
missing_grade_df.dropna()
```

<br><br>
Basic aggregate statistics are affected by dropping nulls/nas.

```python
missing_grade_df.groupby('grade')[['reading_score','math_score']].agg(['count','sum','mean'])
```

```python
missing_grade_df.dropna().groupby('grade')[['reading_score','math_score']].agg(['count','sum','mean'])
```
